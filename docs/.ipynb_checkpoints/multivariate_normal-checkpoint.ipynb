```python
import pymc3 as pm
import scipy.stats as st
from scipy.special import binom
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from context import predicate_search
from predicate_search import RobustNormal, Feature, Predicate, Transform
```

```python
def generate_norm(n, m):
    mean = np.random.normal(0, 10, size=m)
    cov = st.invwishart(scale=np.ones(m)*.01).rvs()
    if m > 1:
        data = np.random.multivariate_normal(mean, cov, size=n)
    else:
        data = np.random.normal(mean, cov, size=n)
    return pd.DataFrame(data)
```

```python
def generate_features(n, m):
    data = np.random.uniform(0, 1, size=(n,m))
    return pd.DataFrame(data)
```

```python
def generate_data(n, m, target_m):
    vals = generate_norm(n, target_m)
    features = generate_features(n, m)
    df = pd.concat([vals.rename(columns={i: f'target_{i}' for i in range(target_m)}),
                    features.rename(columns={i: f'feature_{i}' for i in range(m)})], axis=1)
    return df
```

```python
def discritize(d, bins):
    disc_d = ((d - d.min()) / (d.max() - d.min()) * (bins - 1)).astype(int)
    inv_idx_map = pd.Series(sorted(disc_d.unique()))
    idx_map = pd.Series(inv_idx_map.index, index=inv_idx_map.values)
    disc_d = disc_d.map(idx_map)
    return disc_d

def discritize_df(df, bins, target_cols, discrete_cols=[]):
    disc_data = df.copy()
    dtypes = {}
    for col in disc_data.columns:
        if col not in discrete_cols and col not in target_cols:
            disc_data[col] = discritize(disc_data[col], bins)
    return disc_data
```

```python
def generate_predicate(data, disc_data, m):
    cols = ['feature_' + str(i) for i in np.random.choice(np.arange(int(data.columns[-1].split('_')[-1])), size=m)]
    features = []
    for col in cols:
        idx = np.random.choice(np.arange(4))
        val_min, val_max = np.quantile(data[col], [0, .25, .5, .75, 1])[[idx, idx+1]]
        disc_vals = disc_data[col][(data[col] >= val_min) & (data[col] <= val_max)]
        disc_min = disc_vals.min()
        disc_max = disc_vals.max()
        values = np.arange(disc_min, disc_max)
        data_idx = disc_data[disc_data[col].isin(values)].index.tolist()
        feature = Feature(col, values, data_idx, True)
        features.append(feature)
    predicate = Predicate(features, np.zeros(data.shape[0]))
    return predicate
```

```python
def insert_anomalies(data, disc_data, anomalies, predicate):
    query = ' and ' .join(
        [f"{f.feature} >= {f.values[0]} and {f.feature} <= {f.values[-1]}"
         for f in predicate.features])
    idx = disc_data.query(query).index.tolist()
    
    tainted_data = data.copy()
    tainted_data.iloc[idx] = anomalies.iloc[idx]
    return tainted_data, predicate, idx
```

```python
class PredicateSearch:

    def __init__(self, df, dtypes, scores, target):
        self.df = df.copy()
        self.dtypes = dtypes
        self.scores = scores
        self.target = target
        self.base_predicates = self.generate_base_predicates(scores)

    def generate_base_predicates(self, scores):
        p = [[Predicate(
            [Feature(feature, [value], self.df[self.df[feature] == value].index.tolist(),
                     self.dtypes[feature] == 'continuous')], scores) for value in sorted(self.df[feature].unique())]
                     for feature in self.df.columns]
        p = [a for b in p for a in b]
        return p

    def set_scores(self, predicates, c):
        for p in predicates:
            p.set_score(c)

    def merge_adjacent(self, predicates, idx, c, eps):
        offset = 0
        predicate = predicates[idx]
        for i in range(idx + 1, len(predicates)):
            if predicate.is_adjacent(predicates[i - offset]):
                merged_predicate = predicate.merge(predicates[i - offset])
                merged_predicate.set_score(c)
                if (merged_predicate.score - eps) <= predicate.score:
                    predicates[idx] = merged_predicate
                    predicate = merged_predicate
                    del predicates[i - offset]
                    offset += 1

    def merge_all_adjacent(self, predicates, c, eps):
        idx = 0
        while idx < len(predicates):
            self.merge_adjacent(predicates, idx, c, eps)
            idx += 1
        return predicates

    def merge_all_all_adjacent(self, predicates, c, eps, maxiters=100):
        for i in range(maxiters):
            n = len(predicates)
            predicates = self.merge_all_adjacent(predicates, c, eps)
            new_n = len(predicates)
            if n == new_n:
                return predicates

    def prune(self, predicates, best_score):
        return [p for p in predicates if p.best_score - 10 ** -5 < best_score]

    def intersect(self, predicates):
        new_predicates = [p1.merge(p2) for p1, p2 in itertools.combinations(predicates, 2) if not p1.is_adjacent(p2)]
        new_predicates = [p for p in new_predicates if p.raw_score < np.inf]
        return new_predicates

    def search(self, c=1., eps=.01, maxiters=10, idx=None):
        best_score = np.inf
        best_p = []
        if idx is None:
            predicates = self.base_predicates.copy()
        else:
            predicates = np.array(self.base_predicates)[
                [np.array(idx)[:,None] in p.data_idx for p in self.base_predicates]].tolist()
        SMALL_NUM = 10**-5

        for i in range(maxiters):
            self.set_scores(predicates, c)
            used_features = list(set([a for b in [p.feature_names for p in best_p] for a in b]))
            predicates = [p for p in predicates if (p.score < best_score - SMALL_NUM) or
                          (p.score - SMALL_NUM < best_score and not any(i in p.feature_names for i in used_features))]
            if len(predicates) == 0:
                return best_p
            print(i, len(predicates))
            
            sorted_predicates = sorted(predicates, key=lambda x: x.score)
            merged_predicates = self.merge_all_all_adjacent(sorted_predicates, c, eps)

            new_best_p = min(merged_predicates, key=lambda x: x.score)
            new_best_score = new_best_p.score
            if new_best_score < best_score - SMALL_NUM:
                best_score = new_best_score
                best_p = [new_best_p]
            elif new_best_score - SMALL_NUM < best_score:
                best_p.append(new_best_p)
            else:
                return best_p

            best_single_score = min([p.best_score for p in best_p])
            pruned_predicates = self.prune(merged_predicates, best_single_score)
            predicates = self.intersect(pruned_predicates)
        return best_p
    
    def search_all(self, max_runs=3, c=1., eps=.01, maxiters=10, idx=None):
        mean = np.mean(self.target)
        std = np.std(self.target)
        min_val, max_val = st.norm(mean, std).interval(.8)
        norm = np.mean([t >= min_val and t <= max_val for t in self.target])
        x = np.linspace(self.target.min(), self.target.max())
        f = st.norm(mean, std).pdf(x)
        print(min_val, max_val, norm)

        fig, ax = plt.subplots()
        pd.Series(self.target).hist(bins=800, density=1, ax=ax)
        ax.plot(x, f)
        plt.show()
        
        predicates = []
        for i in range(max_runs):
            new_p = self.search(c=c, eps=eps, maxiters=maxiters, idx=idx)
            predicates += new_p
            for p in new_p:
                q = ' and '.join([f'{f.feature} in {f.values}' for f in p.features if f])
                self.df = self.df.query('~(' + q +')')
            self.scores = self.scores[self.df.index]
            self.target = self.target[self.df.index]
            self.df = self.df.reset_index(drop=True)
            
            mean = np.mean(self.target)
            std = np.std(self.target)
            min_val, max_val = st.norm(mean, std).interval(.8)
            norm = np.mean([t >= min_val and t <= max_val for t in self.target])
            x = np.linspace(self.target.min(), self.target.max())
            f = st.norm(mean, std).pdf(x)
            print(min_val, max_val, norm, new_p[0].score)
            
            fig, ax = plt.subplots()
            pd.Series(self.target).hist(bins=800, density=1, ax=ax)
            ax.plot(x, f)
            plt.show()
            self.base_predicates = self.generate_base_predicates(self.scores)
            #print(max([p.raw_score for p in self.base_predicates]))
        return predicates
```
